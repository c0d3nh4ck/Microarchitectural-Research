```python
import glob
import statistics as st
import numpy as np
```

```python
uops_set = {
    '0' : [1],
    '1' : [1, 2, 3, 4, 5, 6, 8, 9, 10, 11, 12, 13, 14, 15, 16],
    '16': [1],
    '17': [1],
    '18': [1],
    '2' : [1, 2, 3, 4, 5, 7, 12, 14, 15, 16],
    '20': [1],
    '22': [1],
    '25': [1],
    '27': [1],
    '3' : [2, 7],
    '31': [1],
    '34': [1],
    '4' : [1],
    '5' : [1],
    '8' : [1],
    '89': [1],
    '9' : [1]
}
```

```python
values_dict = {}
for i in list(uops_set.keys()):
    for j in uops_set.get(i):
        path = "timing_data/"
        path += "time_" + i + "/time_ah_" + i + "_" + str(j) + "_*"

        files_list = glob.glob(path)
        arr_of_avg = []
        dict_key = i + '_' + str(j)

        for filename in files_list:
            # parse the files to calculate average of timing values for each file
            # store the timing values in a dictionary as mentioned in the rough page
            with open(filename, 'r') as f:
                single_file_data = f.readlines()

            arr_of_single_file = []
            for elem in single_file_data:
                arr_of_single_file.append(int(elem))

            arr_of_avg.append(st.mode(np.array(arr_of_single_file)))

        values_dict[dict_key] = arr_of_avg
```

```python
def lsq(x1, x2):
    res = 0
    for i in range(20):
        res += pow((x1[i] - x2[i]), 2)

    return pow(res/20, 0.5)
```

```python
lsq_dict = {}
lsq_values2 = 99999
for key in list(values_dict.keys()):
#     print(key)
    lsq_values1 = lsq(values_dict.get(key), arr_of_target_avg)
    lsq_dict[key] = lsq_values1

    if lsq_values1 < lsq_values2:
        lsq_values2 = lsq_values1
        instr_set = key
```
